```python
import pandas as pd
df_allergies = pd.read_csv ('/content/allergies.csv')
```

```python
import pandas as pd
df_patients = pd.read_csv ('/content/patients.csv')
```

```python
dict={}
l=df_patients.shape[0]
for i in range(0,l):
  dict[df_patients['Id'][i]]=df_patients['COUNTY'][i]
print(dict)
```

```python
dict1={}
l1=df_allergies.shape[0]
for i in range(0,l1):
  county=dict[df_allergies['PATIENT'][i]]
  allergy=df_allergies['DESCRIPTION'][i]
  if county not in dict1:
    dict1[county]={}
  dict2=dict1[county]
  if allergy not in dict2:
    dict2[allergy]=1
  else:
    dict2[allergy]+=1
print(dict1)
```

```python
df = pd.DataFrame(dict1)
print(df)

df.transpose().plot.bar(figsize =(20, 8),width=0.65)
```

```python
dict3={}
l=df_patients.shape[0]
for i in range(0,l):
  dict3[df_patients['Id'][i]]=df_patients['GENDER'][i]
print(dict3)
```

```python
dict4={}
l1=df_allergies.shape[0]
for i in range(0,l1):
  gender=dict3[df_allergies['PATIENT'][i]]
  allergy=df_allergies['DESCRIPTION'][i]
  if gender not in dict4:
    dict4[gender]={}
  dict5=dict4[gender]
  if allergy not in dict5:
    dict5[allergy]=1
  else:
    dict5[allergy]+=1
print(dict4)
```

```python
df4 = pd.DataFrame(dict4)
print(df4)

df4.transpose().plot.bar(figsize =(20, 8),width=0.65)
```

```python
dict6={}
l=df_patients.shape[0]
for i in range(0,l):
  dict6[df_patients['Id'][i]]=df_patients['BIRTHDATE'][i]
print(dict6)
```

```python
from datetime import datetime
date_format = "%m/%d/%Y"
dict7={}
l1=df_allergies.shape[0]
for i in range(0,l1):
  dob=dict6[df_allergies['PATIENT'][i]]
  allergy=df_allergies['DESCRIPTION'][i]
  start_date=df_allergies['START'][i]
  a = datetime.strptime(start_date, date_format)
  b = datetime.strptime(dob, date_format)
  delta = a-b
  if allergy not in dict7:
    dict7[allergy]=[]
  dict7[allergy].append(delta.days/365)
print(dict7)
```

```python
import matplotlib.pyplot as plt
nkeys=len(dict7.keys())
fig, axs = plt.subplots(nrows=3, ncols=5)
i=0
j=0
for key, data in dict7.items():
  axs[i][j].hist(data)
  axs[i][j].set_title(key)
  j+=1
  if(j==5):
    i+=1
    j=0
  plt.rcParams["figure.figsize"] = (23,8)
plt.legend(loc='upper right')
plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)
plt.suptitle("Age(in years)")
plt.show()
```
